```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

# Now we will try to find the global optimum solution using hill climbing with random restarts

def func(x):
    return x**5-10*x**3+30*x

x=np.linspace(-3,3,200)
plt.plot(x,func(x))
```

```python
import random
def random():
    import random
    
        

    
    p=1
    MAX_VALUE=3
    MIN_VALUE=-3
    ls=[]
    ls1=[]
    
    
    S=random.random()
    best=S
    T=random.uniform(-1,1)
    t=random.choice(T)
    while(t>0):
        
        while(True):

            while(True):

                n=random.randrange(-15,19)
                if S+n<=MAX_VALUE and S+n>=MIN_VALUE:
                    ls.append(S)
                    ls1.append(best)
                    
                    
                    break
            x1=S
            S+=n
            R=S
            if func(R)>func(x1):
                S=R
                continue
            else:
                break
        
        if func(S)>func(best):
            best=S
            continue
        else:
            break
        t=t-(random.uniform(-1,1))
        
        
        
    return ls+ls1

print(random())
```
