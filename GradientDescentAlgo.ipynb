```python
# learning rate
alpha = 1e-3 
# change it to 1e-6 and 1e3

# local minima is at w=8/3=2.66
# gradient computation
def df(w):
    return 2 * w**3 - 11 * w**2

def convergence_not_reached(step_sz):
    precision = 1e-6
    return abs(step_sz) > precision

iter   = 1
w      = 1
w_next = 0.9
while convergence_not_reached(w_next-w):
    print ("iteration=", iter, " w=", w, " and w_next", w_next)
    iter += 1
    
    w      = w_next
    w_next = w - alpha * df(w)

    
print("GD found minima at", "{:.3f}".format(w))
```
