```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

# Now we will try to find the global optimum solution using hill climbing with random restarts
import math
def func(x):
    return x**3+np.log(x**2)+x**4*np.sin(x**3)

x=np.linspace(-3,3,200)
plt.plot(x,func(x))
```
